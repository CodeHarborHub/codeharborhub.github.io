<h3>Bias and Weight Visualization in Multiple Regression</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def linear_model(x , w , b):
    return w*x + b

weights = np.array([0 ,2 ,3 ,4])
biases = np.array([-1, 0.5 , -0.5 , 2])

b = 1
w = 1
x = np.linspace(-10 , 10 , 200)

for weight in weights:
    y = linear_model(x , weight , b)
    plt.plot(x , y  , label = f"y = {weight}x + {b}")

plt.title("Effets of Weight")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.grid(True)
```

```python
for bias in biases:
    y = linear_model(x , w , bias)
    plt.plot(x , y  , label = f"y = {w}x + {bias}")

plt.title("Effets of Bias")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.grid(True)
```
